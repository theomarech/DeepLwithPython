# CHPT III - Un exemple de DNN pour la régression : exemple du jeux de données Boston

- normalisation des données quantitative : importance de centrée réduire les prédicteurs!!!!!
- jeux de données de petite taille : 506 404/102
- un exemple de régression

objectif prédire les prix des maisons à boston en fonction de leurs caractéristiques

```python
from keras.datasets import boston_housing

(train_data,train_labels),(test_data,test_labels) = boston_housing.load_data()

train_data.shape # 13 colonnes
type(train_data)
```

## Préaparation des données
normalisation

```python
train_data -= train_data.mean(axis=0)
train_data /= train_data.std(axis=0)

test_data -= test_data.mean(axis=0)
test_data /= test_data.std(axis=0)
```

## architecture du réseaux

```python
from keras import layers, models

def build_model():
    model = models.Sequential()
    model.add(layers.Dense(64,activation="relu",input_shape=(13,)))
    model.add(layers.Dense(64, activation="relu"))
    model.add(layers.Dense(1))
    model.compile(optimizer = "rmsprop",
                 loss="mse",
                 metrics = ["mae"])
    return model

```

## Définition d'une fonction Kfold

```python
import numpy as np
k = 4
size = len(train_data)
index = np.arange(size)
fold_size = size//k
model_eval = []
for i in range(k):
    print("processing fold #", i+1)
    val_data = train_data[i * fold_size : (i+1) * fold_size]
    val_targets = train_labels[i * fold_size : (i+1) * fold_size]
    
    partial_train_data = np.concatenate([train_data[:i*fold_size],
                                         train_data[(i+1) * fold_size:]],axis=0)
    partial_train_targets = np.concatenate([train_labels[:i*fold_size],
                                         train_labels[(i+1) * fold_size:]],axis=0)
    
    model = build_model()
    
    model.fit(partial_train_data,
             partial_train_targets,
             epochs = 500,
             batch_size = 1,
             validation_data = (val_data,val_targets),
             verbose = 0)
    mae_history = model.history.history["val_mae"]
    model_eval.append(mae_history)

```

## Courbe d'apprentissage

On va observer ici comment la courbe de test évolue pour identifier les hyperparamètres, pour ça on va:
- utiliser une courbe simple en fonction des epochs de la moyenne des mae en fonction des folds pour chaque epochs
- utiliser une [moyenne mobile exponentielle](https://fr.wikipedia.org/wiki/Moyenne_mobile) pour lisser la courbe
- supprimer les premiers points pour mieux visualiser le minimum local

### visualisation de la courbe brute

```python
mean_mae = [np.mean([x[i] for x in model_eval]) for i in range(500)]    
dic = dict(zip(np.arange(500),mean_mae))
rev_dic = dict([(v,k) for (k,v) in dic.items()])
min_mae_epoch = rev_dic.get(np.min(mean_mae))
import matplotlib.pyplot as plt

plt.figure(figsize = (15,10))
plt.plot(range(500),mean_mae,label="courbe d'apprentissage sur les folds")
plt.axvline(min_mae_epoch,linestyle="--",alpha=0.5)
plt.annotate("mae minimum de "+str(round(np.min(mean_mae),2))+" à l'epoch "+str(min_mae_epoch),(min_mae_epoch,3.5))
plt.xlabel('nombre d\'epochs')
plt.ylabel('erreur moyenne absolue (mae)')
plt.title('courbe d\'apprentissage: mae en fonction du nombre d\'epoch')
```

```python
def smooth_curve(data,factor=0.9):
    """fonction qui calcul la moyenne mobile exponentielle. prend en entrée une liste,\
    un array et retourne une liste contenant les moyennes mobile exponentielle"""
    exp_point = []
    for i in range(len(data)):
        if i == 0:
            exp_point.append(data[i])
        else:
            exp_point.append(exp_point[-1]*factor + data[i]*(1-factor))
    return(exp_point)

def smooth_curve2(points,factor=0.9):
    """voir doc de la fonction smooth_curve"""
    smoothed_points = []
    for point in points:
        if smoothed_points: # si une liste est vide retourne False sinon True ! pas mal comme comportement
            previous = smoothed_points[-1] # prend le dernier points de smoothed
            smoothed_points.append(previous * factor + point * (1-factor)) # calcule de la moyenne mobile
        else:
            smoothed_points.append(point)
    return(smoothed_points)
        
```

```python
smooth_mae = smooth_curve(mean_mae[10:])

plt.figure(figsize = (15,10))
plt.plot(range(490),smooth_mae,label="courbe d'apprentissage sur les folds")
plt.xlabel('nombre d\'epochs')
plt.ylabel('erreur moyenne absolue (mae)')
plt.title('courbe d\'apprentissage: mae en fonction du nombre d\'epoch')
```

On observe qu'il n'y a plus d'augmentation significative après la 80 ème epochs. On pourra optimiser le modèle en réglant les autres hyperparamètres telle que le nombre de couche cachées, le nombre d'unités par couches, puis re faire la meme chose.

```python
model = build_model()

model.fit(train_data,train_labels,epochs=80,batch_size=16,verbose=0)

mse_score, mae_score = model.evaluate(test_data,test_labels)
```

## Comparaison avec un modèle linéaire classique et un random forest

```python
from sklearn.linear_model import ElasticNet
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_absolute_error

param = {"l1_ratio":np.logspace(-4,0)}

lin = ElasticNet()

grid_lin = GridSearchCV(lin,param,cv=5)

grid_lin.fit(train_data,train_labels)

grid_lin.score(train_data,train_labels)

pred_y_test = grid_lin.predict(test_data)

print(grid_lin.best_param)

mean_absolute_error(test_labels,pred_y_test)
```
