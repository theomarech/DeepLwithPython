# Architecture du jeux de données

````python
from keras import models,layers

model = models.Sequential()
model.add(layers.Conv2D(32,(3,3),activation="relu",input_shape = (150,150,3)))
model.add(layers.MaxPooling2D((2,2)))
model.add(layers.Conv2D(64,(3,3),activation="relu"))
model.add(layers.MaxPooling2D((2,2)))
model.add(layers.Conv2D(128,(3,3),activation="relu"))
model.add(layers.MaxPooling2D((2,2)))
model.add(layers.Conv2D(128,(3,3),activation="relu"))
model.add(layers.MaxPooling2D((2,2)))
model.add(layers.Flatten())
model.add(layers.Dense(512,activation="relu"))
model.add(layers.Dense(1,activation="sigmoid"))

model.summary()
````

````python
from keras import optimizers

model.compile(optimizer = optimizers.RMSprop(lr=1e-4),
             loss = "binary_crossentropy",
             metrics = ["acc"])
````

Nous allons utiliser un générateur de minilots à l'aide de la classe```ImageDataGenerator``` et de la méthode ```flow_from_directory```contenue dans ```keras.preprocessing.image```. 
Le générateur permet de créer des lot dont les caractéristiques sont standardisées ou justement transformer aléatoirement.

## Use pre-trained models to improve performance in computer vision
In this part we will use **VGG16** a pre-trained model which was trained in 2014 on **ImageNet** database for image recognition. 
Like we have seen previously a ConvNet is build with two part. One part consist in different layers sequence alterning convolution and pooling names **"Convultion basis"**, the other part consist in a **classifier** build with dense layers.
The first part learn spatial hierarchical features which can be transposable from one learning task to another. On the other hand, the classifier is more restrictive and learn the specificity of the pictures class on which it was trained and break spatial caracteristic. The trick is to use the convolution basis and probably more the first layers of this part which have learned more generic caracterstics map.
In our special case of classification between cats and dogs with **VGG16** model, we could used all the pretrained model because it was trained on dogs and cats pictures but for the training we will use two strategy as if **VGG16** have not been trained on  our classification task.

### Instantiation de la base de convolution

````python
from keras.applications import VGG16

conv_base = VGG16(weights = "imagenet", # les poids issues de l'entrainement sur imagenet
                  include_top = False, # on ne veut pas la couche dense 
                  input_shape = (150,150,3)) # la taille de nos images : OPTIONEL

conv_base.summary()
````

On peut voir que la base de convolution est constituée de successions de plusieurs couches de convolution de même taille suivit de maxpooling. Au début ces couches s'enchainent par deux puis par trois avec au total 5 enchainements (2x64;2x128;3x256;3x512 x 2) pour un total d'environs 15 millions de paramètres. On peut observer que le maxpooling se fait avec un une fenetre de 2x2 divisant systématiquement la taille des cartes de caractéristiques par 2.

Deux manières de faire ensuite pour utiliser cette base de convolution :

- la faire tourner avec notre jeux de données puis enregistrer la sortie dans un tableau numpy pour entrainer un DNN ensuite dessus. avantage : très rapide, désavantage : pas possible d'utiliser de l'augmentation de données et donc résultats moins bon
- utiliser l'augmentation de données et donc entrainer de bout en bout, entièrement, en ajoutant à notre base de convolution un DNN. => pas possible sans GPU :'( mais résultats bien meilleur

### Extraction rapide des caractéristiques sans augmentation des données

On crée  une fonction qui va extraire les caractéristiques de sortie de la base de convolution de VGG16. Pour ça on va créer un générateur pour utiliser nos images à partir de notre dossier puis les faire passer dans la base de convolution et récupérer les données dans des tableaux numpy.

````python
import os
import numpy as np
from keras.preprocessing.image import ImageDataGenerator

base_dir =  "/home/theo/Desktop/auto_formation_data/deeplwithpython/cats_and_dogs_small"
train_dir = os.path.join(base_dir,"train")
validation_dir = os.path.join(base_dir,"validation")
test_dir = os.path.join(base_dir,"test")



datagen = ImageDataGenerator(rescale = 1./255)

batch_size = 20

def extract_features(directory,sample_count):
    """Extrait les caractéristiques à partir de la base de convolution VGG16
    Args:
        directory = chemin du dossier contenant des images
        sample_count = nombre d'images souhaitée (<= au nombre total d'images présentent dans le dossiers)
    Return
        deux arrays numpy"""
    features = np.zeros(shape=(sample_count,4,4,512))
    labels = np.zeros(shape=(sample_count))
    generator = datagen.flow_from_directory(directory, 
                                            target_size = (150,150), 
                                            batch_size = batch_size, 
                                            class_mode ="binary")
    i = 0
    
    for input_batch, label_batch in generator:
        features_batch = conv_base.predict(input_batch) #on prédit
        
        features[i*batch_size : (i+1) * batch_size] = features_batch # on récupère la prédiction
        labels[i*batch_size : (i+1) * batch_size] = label_batch # on récupère les labels
        
        i+=1
        
        if i * batch_size >= sample_count:
            break
    return features, labels


train_features, train_labels = extract_features(train_dir, sample_count = 2000)
validation_features, validation_labels = extract_features(validation_dir, sample_count = 1000)
test_features, test_labels = extract_features(test_dir, sample_count = 1000)

````

On reshape les tenseurs pour qu'ils puissent rentrer dans notre réseaux de neurones dense en les passant de la forme (n,h,w,c) en (n,h*w*c). On applatit la dernière couche de convolution en d'autres termes !

````python
train_features = np.reshape(train_features,(2000,4*4*512))
validation_features = np.reshape(validation_features,(1000,4*4*512))
test_features = np.reshape(test_features,(1000,4*4*512))
````

Création du DNN peu profond avec deux couches.

````python
from keras import layers
from keras import models
from keras import optimizers

model = models.Sequential()
model.add(layers.Dense(256,activation ="relu", input_dim = 4*4*512))
model.add(layers.Dropout(0.5))
model.add(layers.Dense(1, activation ="sigmoid"))
model.compile(optimizer = optimizers.RMSprop(lr=1e-5), 
             loss = "binary_crossentropy",
             metrics = ["acc"])

model.fit(train_features,
          train_labels,
          batch_size = 20,
          epochs = 30,
          validation_data = (validation_features, validation_labels))
````

````python
import matplotlib.pyplot as plt

history = model.history.history
loss = history["loss"]
val_loss = history["val_loss"]
acc = history["acc"]
val_acc = history["val_acc"]
epochs = np.arange(1,31)

fig, (ax1, ax2) = plt.subplots(1,2,figsize = (25,7)) 

ax1.plot(epochs, loss, color = "lightblue", marker = "o", linestyle = "none", label = "loss train")
ax1.plot(epochs, val_loss, color = "darkred", label = "loss validation")
ax1.legend()

ax2.plot(epochs, acc, color = "lightblue", marker = "o", linestyle = "none", label = "accuracy train")
ax2.plot(epochs, val_acc, color = "darkred", label = "accuracy validation")
ax2.legend()

fig.suptitle("courbes d'entrainements et de validaitons pour la perte et l'exactitude pour un ConvNet pré entrainé à partir de la base de convolution (VGG16) avec une tête DNN nouvelle")
````

On observe un bien meilleur score pour notre modèle avec une accuracy proche de 90% avec un modèle qui a tourné très vite comparé à notre ConvNet crée à partir de zéros. Contrairement à ce qui est dit dans le livre le sur-ajustement est assez faible ici bien qu'on atteigne rapidement une accuracy max (~10 epochs?). 

## Extraction de caractéristiques avec augmentation de données
C'est la meilleur manière d'obtenir un classifieur très efficace, on combine ici notre réseaux pré-entrainé avec de l'augmentation de données. Il faut donc re faire passer nos données sur l'ensemble du réseau préentrainé avec un nouveau classifieur peu profond (le même que précédemment). Le problème est que cette technique est fortement intense en temps de calcul. Sur un GPU c'est faisable mais pas sur un CPU. Le code suivant est donc là pour illustration mais ne sera pas executé sans s'assurer qu'il sera executé sur un GPU sans quoi on pourra rapidement arrêter le kernel du notebook !

### Ajout d'un classifieur entièrement connectée à la base de convolution de VGG16

````python
from keras import layers
from keras import models
from keras import optimizers
from keras.applications import VGG16
from keras.preprocessing.image import ImageDataGenerator
import os
import shutil

# instantiation de la base de convolution
base_conv = VGG16(weights = 'imagenet',
                 include_top = False,
                 input_shape = (150,150,3))

# création des chemins pour accéder aux données
base_dir = '/home/theo/Desktop/auto_formation_data/deeplwithpython/cats_and_dogs_small'
train_dir = os.path.join(base_dir,"train")
validation_dir = os.path.join(base_dir,'validation')
test_dir = os.path.join(base_dir,'test')

# réseaux de neurones

model = models.Sequential()
model.add(base_conv)
model.add(layers.Flatten())
model.add(layers.Dense(256, activation = 'relu'))
model.add(layers.Dense(1, activation = 'sigmoid'))

model.summary()
````

On peut observer le très grand nombre de paramètre (presque 40 fois plus que notre convnet de base) qui a mis déjà une bonne demie-heure à tourner sur un laptop sans GPU.

Avant de compiler le modèle et de le lancer il est très important (nécessaire) de **figer la base de convolution**. Figer une couche signifie simplement empecher la MàJ des poids de cette couche pendant l'entrainement.

````python
print("this is the number of trainable weights:", len(model.trainable_weights))
base_conv.trainable = False # on fige la base de convolution
print("this is the number of trainable weights after freezing:", len(model.trainable_weights))
````

Ainsi seulement les poids des deux couches denses seront entrainés soit 2 tenseurs par couches (les poids et les biais) = 4 tenseurs à entrainer.
Maintenant il faut compiler le modèle:

````python
model.compile(optimizer = optimizers.RMSprop(lr=2e-5), # comment choisir un bon learning rate ? that is the question => cross validation maybe ?
             loss = 'binary_crossentropy',
             metrics = ["acc"])
````

Préparation des générateurs d'images et des données

````python
# préparation des données pour le générateurs
test_datagen = ImageDataGenerator(rescale = 1./255)

train_datagen = ImageDataGenerator( 
        rescale = 1./255,
        height_shift_range = 0.2,
        width_shift_range = 0.2,
        rotation_range = 40,
        shear_range = 0.2,
        zoom_range = 0.2,
        horizontal_flip = True,
        fill_mode = "nearest") 

# création des générateurs
train_generator = train_datagen.flow_from_directory(
        train_dir,
        target_size = (150,150),
        batch_size = 20,
        class_mode = 'binary')

validation_generator = test_datagen.flow_from_directory(
        validation_dir,
        target_size = (150,150),
        batch_size = 20,
        class_mode ='binary')
````

Et on lance le modèle... 

````python
model.fit(train_generator,
         steps_per_epoch = 100,
         epochs = 30,
         validation_data = validation_generator,
         validation_steps = 50)
````
