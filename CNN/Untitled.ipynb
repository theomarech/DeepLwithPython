## I - Visualisation des cartes de caractéristiques des couches intermédiaires
Objectif est de voir comment les couches transforme les entrées. On va visualiser ici les **activations intermédiaires** ce qui consiste à afficher les **cartes de caractéristiques** qui sont produites par les différentes couches de convolution et de pooling (la sortie d'une couche, donc la sortie de sa fonction d'activation est souvent appelée son **activation**).
Ces cartes sont en trois dimensions (H,W,C). Chaque cannaux doit être visualisé indépendamment (RGB) car ils encondent des caractéristiques relativement indépendantes. 
Ici nous allons utilisé le modèle que nous avons enregistré en deuxième (avec augmentation de données) et regarder les cartes de caractéristiques des couches intermédiaires.

````python
from keras.models import load_model

model = load_model('cat_and_dog_small_2.h5')
model.summary()
````

Maintenant on va prendre une image en entrée sur lequel le réseau ne s'est pas entrainé.

````python
img_path = "/home/theo/Desktop/auto_formation_data/deeplwithpython/cats_and_dogs_small/test/cat/cat.1500.jpg"

from keras.preprocessing import image
import numpy as np

img = image.load_img(img_path,target_size = (150,150))
img_tensor = image.img_to_array(img)
print(img_tensor.shape)
img_tensor = np.expand_dims(img_tensor, axis = 0)
print(img_tensor.shape)
img_tensor /= 255.
print(img_tensor.shape)
````

On affiche l'image : 

````python
import matplotlib.pyplot as plt

plt.imshow(image.array_to_img(img_tensor[0]))
plt.show()
````

Pour extraire les cartes de caractéristiques on va créer un modèle Keras qui prend en entrée des lots d'images et qui affiche en sortie les activations de toutes les couches de convolution et de pooling. Pour ça on va utiliser la classe **Mode1**. Un modèle dans keras est instantié avec deux arguments :
- tenseur d'entrée (ou liste de tenseur d'entrée)
- tenseur de sortie (ou liste)

```.Model()``` fonctionne donc comme le modèle ```.Sequential()``` à la différence qu'il permet des sorties multiples. (voir chapitre 7.1 notebook :)


````python
from keras import models

layer_outputs = [layer.output for layer in model.layers[:10]]
print(layer_outputs)

activation_model = models.Model(inputs = model.input,
                               outputs = layer_outputs)

````

C'est donc un modèle à plusieurs sorties ici avec une entrée, plus précisémment une entrée et 8 sorties. C'est la première fois qu'on voit ca ici mais dans le cas général un modèle peut avoir un nombre indéfini d'entrées et de sorties.

````python
activations = activation_model.predict(img_tensor)
first_layer_activation = activations[0]

plt.matshow(first_layer_activation[0,:,:,7],cmap='viridis')
````

````python
layer_names = []
for layer in model.layers[:8]:
    layer_names.append(layer)

images_per_row = 16

for layer, layer_activation in zip(layer_names, activations):
    print(layer)
````
